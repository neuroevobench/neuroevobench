## Visualize Grid Search Results for Open ES

```python
import matplotlib.pyplot as plt
from mle_toolbox import load_result_logs
from mle_toolbox.visualize import visualize_2D_grid
```

```python
fig, axs = plt.subplots(1, 3, figsize=(25, 10))

meta_log_cart, hyper_log_cart = load_result_logs("../experiments/Open_ES/cartpole")
visualize_2D_grid(
    hyper_log_cart, 
    params_to_plot=['es_params/sigma_init',
                    'es_params/lrate_init'],
    target_to_plot="test_perf_strat_mean",
    plot_title="Open ES on CartPole (LSTM)",
    xy_labels=["sigma_init", "lrate_init"],
    round_ticks=3,
    fig=fig,
    ax=axs[0])

meta_log_ant, hyper_log_ant = load_result_logs("../experiments/Open_ES/ant")
visualize_2D_grid(
    hyper_log_ant, 
    params_to_plot=['es_params/sigma_init',
                    'es_params/lrate_init'],
    target_to_plot="test_perf_strat_mean",
    plot_title="Open ES on Brax-Ant (MLP)",
    xy_labels=["sigma_init", "lrate_init"],
    round_ticks=3,
    fig=fig,
    ax=axs[1])

# meta_log, hyper_log = load_result_logs("../experiments/Open_ES/mnist")
visualize_2D_grid(
    hyper_log_ant, 
    params_to_plot=['es_params/sigma_init',
                    'es_params/lrate_init'],
    target_to_plot="test_perf_strat_mean",
    plot_title="Open ES on MNIST (CNN)",
    xy_labels=["sigma_init", "lrate_init"],
    round_ticks=3,
    fig=fig,
    ax=axs[2])

fig.tight_layout()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(25, 8))
meta_log_cart.plot("test_perf_strat_mean", num_legend_cols=5, fig=fig,
    ax=axs[0], plot_title="Open ES on CartPole (LSTM)")
meta_log_ant.plot("test_perf_strat_mean", num_legend_cols=5, fig=fig,
    ax=axs[1], plot_title="Open ES on Brax-Ant (MLP)")
meta_log_ant.plot("test_perf_strat_mean", num_legend_cols=5, fig=fig,
    ax=axs[2], plot_title="Open ES on MNIST (CNN)")
```
